```python
import pandas as pd
```

## Loading the dataset

```python
df_dt = pd.read_csv('C:\\Users\\rahul\\Data Science\\Material\\Social_Network_Ads.csv')
df_dt.head()
```

## feature column are: Age, EstimatedSalary and Traget Column: Purchased

```python
x = df_dt.iloc[:,2:4].values
```

```python
y = df_dt.iloc[:,-1].values
```

## Training and spliting the Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.25,random_state = 42)
```

## using gini as criterion

```python
from sklearn.tree import DecisionTreeClassifier
model_dtc = DecisionTreeClassifier(criterion = 'gini',random_state = 0, max_depth = 5)
```

## Training the data in Decision tree classifier Model

```python
model_dtc.fit(x_train,y_train)
```

```python
y_pred = model_dtc.predict(x_test)
```

```python
model_dtc.predict([[28,25000]]) #predicting by assigning own data to check in which class it belongs
```

## we want to check the model confusion matrix, accuracy, precision, recall or we can get all this at one place using the classification report.

```python
from sklearn.metrics import classification_report,confusion_matrix
print(f'confusion_matrix:\n',confusion_matrix(y_test,y_pred))
```

```python
print(f'classification_report:\n',classification_report(y_test,y_pred))
```

## using entropy as criterion

```python
from sklearn.tree import DecisionTreeClassifier
model_dtc = DecisionTreeClassifier(criterion = 'entropy',random_state = 0, max_depth = 5)
```

```python
model_dtc.fit(x_train,y_train)
y_pred = model_dtc.predict(x_test)
```

```python
print(f'confusion_matrix:\n',confusion_matrix(y_test,y_pred))
```

```python
print(f'classification_report:\n',classification_report(y_test,y_pred))
```
