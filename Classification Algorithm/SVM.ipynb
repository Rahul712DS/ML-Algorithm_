```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading the dataset

```python
df_svm = pd.read_csv('C:\\Users\\rahul\\Data Science\\Material\\Social_Network_Ads.csv')
df_svm.head()
```

## feature column are: Age, EstimatedSalary and Traget Column: Purchased

```python
x = df_svm.iloc[:,2:4].values
```

```python
y = df_svm.iloc[:,-1].values
```

## Training and spliting the Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state = 42)
```

```python
from sklearn.svm import SVC
model_svm = SVC(kernel='linear')
```

## Training the data in SVM Model

```python
model_svm.fit(x_train,y_train)
```

```python
y_pred = model_svm.predict(x_test)  #making the prediction using trained dataset on unseen dataset(test dataset).
```

```python
model_svm.predict([[28,25000]]) #predicting by assigning own data to check in which class it belongs
```

## we want to check the model confusion matrix, accuracy, precision, recall or we can get all this at one place using the classification report.

```python
from sklearn.metrics import classification_report,confusion_matrix
print(f'confusion_matrix:\n',confusion_matrix(y_test,y_pred))
```

```python
print(f'classification_report:\n',classification_report(y_test,y_pred))
```

## Using Standard Scaler (it is used to standardised the data values)

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
model_svm.fit(x_train,y_train)
```

```python
y_pred = model_svm.predict(x_test)
```

```python
print(f'confusion_matrix:\n',confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(f'classification_report:\n',classification_report(y_test,y_pred))
```
