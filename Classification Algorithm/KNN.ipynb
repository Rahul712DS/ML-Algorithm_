```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading the dataset

```python
df_knn = pd.read_csv('C:\\Users\\rahul\\Data Science\\Material\\Social_Network_Ads.csv')
df_knn.head()
```

## feature column are: Age, EstimatedSalary and Traget Column: Purchased

```python
x = df_knn.iloc[:,2:4].values
```

```python
y = df_knn.iloc[:,-1].values
```

## Training and spliting the Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.25,random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier() #default neighbors is 5
```

## Training the data in KNN Model

```python
model_knn.fit(x_train,y_train)
```

```python
y_pred = model_knn.predict(x_test) #making the prediction using trained dataset on unseen dataset(test dataset).
```

```python
y_pred
```

```python
model_knn.predict([[28,25000]]) #predicting by assigning own data to check in which class it belongs
```

## we want to check the model confusion matrix, accuracy, precision, recall or we can get all this at one place using the classification report.

```python
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score,precision_score,classification_report
print(f'confusion_matrix:\n',confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(f'classification_report:\n',classification_report(y_test,y_pred))
```

## Using desired neighbors

```python
model_knn = KNeighborsClassifier(n_neighbors=3) #default neighbors is 5
```

```python
model_knn.fit(x_train,y_train)
```

```python
y_pred = model_knn.predict(x_test)
from sklearn.metrics import classification_report
print(f'classification_report:\n',classification_report(y_test,y_pred))
```

## Using Standard Scaler (it is used to standardised the data values)

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
model_knn = KNeighborsClassifier(n_neighbors=5)
```

```python
model_knn.fit(x_train,y_train)
```

```python
y_pred = model_knn.predict(x_test)
from sklearn.metrics import classification_report
print(f'classification_report:\n',classification_report(y_test,y_pred))
```
