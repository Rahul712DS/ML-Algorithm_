```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading the dataset

```python
df_lr = pd.read_csv('C:\\Users\\rahul\\Data Science\\Material\\Social_Network_Ads.csv')
df_lr.head()
```

## feature column are: Age, EstimatedSalary and Traget Column: Purchased

```python
x = df_lr.iloc[:,2:4].values
```

```python
y = df_lr.iloc[:,-1].values
```

## Training and spliting the Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.25,random_state = 42)
```

## Scatter plot of training data, to visualyze the data distribution

```python
plt.scatter(x_train[:,1],y_train, color = 'red')
plt.title('Training Dataset')
plt.xlabel('x-axis: Age|EstimatedSalary')
plt.ylabel('y-axis: Purchased')
plt.show()
```

## Scatter plot of test dataset, to visualyze the data distribution

```python
plt.scatter(x_test[:,1],y_test, color = 'blue')
plt.title('Test Dataset')
plt.xlabel('x-axis: Age|EstimatedSalary')
plt.ylabel('y-axis: Purchased')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
```

## Training the data in Logistic Regression Model

```python
model_lr.fit(x_train,y_train)
```

```python
y_pred = model_lr.predict(x_test) #making the prediction using trained dataset on unseen dataset(test dataset).
```

```python
y_pred
```

## we want to check the model confusion matrix, accuracy, precision, recall or we can get all this at one place using the classification report.

```python
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score,precision_score,classification_report
print(f'confusion_matrix:\n',confusion_matrix(y_test,y_pred))
```

```python
print(f'accuracy_score:',accuracy_score(y_test,y_pred))
```

```python
print(f'precision_score:',precision_score(y_test,y_pred))
```

```python
print(f'recall_score:',recall_score(y_test,y_pred))
```

```python
print(f'classification_report:\n',classification_report(y_test,y_pred))
```
